#### arrange

`numpy.arrange`

returns an ndarray object containing evenly spaced values within a given range.



```python
numpy.arange(start, stop, step, dtype)

#start -> The start of an interval. default is 0
#stop -> The end of an interval
#step -> Spacing between values, default is 1
#dtype -> Data type of resulting ndarray
```

```python
x = np.arange(2,15,3) 
print(x)
```

#### linspace

`numpy.linspace`

function is similar to arange() function. In this function, instead of step size, the number of evenly spaced values between the interval is specified.

```python
numpy.linspace(start, stop, num)

# start -> starting value of the sequence
# stop -> end value of the sequence
# num -> number of evenly spaced samples to be generated. Default is 50
```

```python
x = np.linspace(start = 10, stop = 20, num =10) 
print (x)
```

```python
x = np.linspace(10,20,5, retstep = True)  # return step is set to True
print (x)
```

###### Numpy Broadcasting

* ability of NumPy to treat arrays of different shapes during arithmetic operations.
* If the dimensions of two arrays are dissimilar, element-to-element operations are not possible.
* The smaller array is **broadcast** to the size of the larger array so that they have compatible shapes.

```python
# Ex 1 - Same dimension

a = np.array([1,2,3,4]) 
b = np.array([10,20,30,40]) 
c = a * b 
print (c)
```

```python
#Ex 2 - Different dimension

a = np.array([1,2,3,4]) 
b = np.array([[10,20,30,40],[100,200,300,400]]) 
c = a * b 
print (c)
```

```python
# Ex 3 - Different dimension

a = np.array([[0.0,0.0,0.0],[10.0,10.0,10.0],[20.0,20.0,20.0],[30.0,30.0,30.0]]) 
b = np.array([1.0,2.0,3.0]) 
print(a)
print(b)
print (a + b)
```

#### split

#### append
* This function adds values at the end of an input array. 
* The append operation is not inplace, a new array is allocated. 
* Also the dimensions of the input arrays must match otherwise ValueError will be generated.

`numpy.append(arr, values, axis)`

* values - To be appended to arr. It must be of the same shape as of arr.
* axis - The axis along which append operation is to be done. If not given, both parameters are flattened

```python
a = np.array([[1,2,3],[4,5,6]]) 
print (a) 
print(a.shape)

print (np.append(a, [7,8,9])) 

print (np.append(a, [[7,8,9]],axis = 0) )
print (np.append(a, [[5,5,5],[7,8,9]],axis = 1))
print(np.append(a, [[5,5,5],[7,8,9]],axis = 1).shape)
```

* The FFT stands for Fast Fourier Transformation which is an algorithm for computing DFT. DFT is a mathematical technique which is used in converting spatial data into frequency data.

```python
from matplotlib import pyplot as plt
import numpy as np 
import seaborn as sns
sns.set_style("darkgrid")
#Frequency in terms of Hertz
fre  = 10
#Sample rate
fre_samp = 100
t = np.linspace(0, 2, 2 * fre_samp, endpoint = False )
a = np.sin(fre  * 2 * np.pi * t)
plt.plot(t, a)
plt.xlabel('Time (s)')
plt.ylabel('Signal amplitude')
plt.show()
```

```python
from scipy import fftpack
A = fftpack.fft(a)
frequency = fftpack.fftfreq(len(a)) * fre_samp
plt.stem(frequency, np.abs(A),use_line_collection=True)
plt.xlabel('Frequency in Hz')
plt.ylabel('Frequency Spectrum Magnitude')
plt.show()
```
